# Linear Regrassion


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('headbrain11.csv')

sns.regplot(
    x='Head Size(cm^3)', 
    y='Brain Weight(grams)', 
    data=df
)

plt.title('Regression Plot: Head Size vs Brain Weight')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Brain Weight(grams)')

plt.show()

print(df.head())
```

```python
df.columns = ['HeadSize', 'BrainWeight']
print(df.head())
```

```python
model = smf.ols('BrainWeight ~ HeadSize', data=df).fit()
print(model.summary())
```
