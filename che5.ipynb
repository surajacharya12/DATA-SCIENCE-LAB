```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

# Line Plots

A line plot is a type of plot that displays the relationship between two variables. In Matplotlib, we can create a line plot using the plot() function.

```python

plt.plot([1, 2, 3], [4, 5, 6], marker='o')
plt.title("line plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
Marker="0"
plt.show()
```

# Bar Plots

A bar plot is a type of plot that displays the relationship between two variables. In Matplotlib, we can create a bar plot using the bar() function.

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.bar(x, y)
plt.show()
```

# Histograms

A histogram is a type of plot that displays the distribution of a variable. In Matplotlib, we can create a histogram using the hist() function.

```python

x = np.random.randn(1000)
plt.hist(x, bins=30)
plt.show()
```

# Scatter Plots

A scatter plot is a type of plot that displays the relationship between two variables. In Matplotlib, we can create a scatter plot using the scatter() function

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.show()
```

# Multi-Panel Subplots

A multi-panel subplot is a type of plot that displays multiple plots in a single figure. In Matplotlib, we can create a multi-panel subplot using the subplots() function.

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot([1, 2, 3, 4, 5])

axs[0, 1].bar([1, 2, 3, 4, 5], [5, 4, 3, 2, 1]) 

axs[1, 0].hist(np.random.randn(1000), bins=30)

axs[1, 1].scatter([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])

plt.show()
```

# Box Plots

A box plot is a type of plot that displays the distribution of a variable. In Seaborn, we can create a box plot using the boxplot() function.

```python
sns.boxplot(x=[1, 2, 3, 4, 5])
plt.show()
```

# Pair Plots

A pair plot is a type of plot that displays the relationship between two variables. In Seaborn, we can create a pair plot using the pairplot() function.

```python
df = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [1, 4, 9, 16, 25]
})

# Create pairplot
sns.pairplot(df)
plt.show()
```

# Heat Maps

A heat map is a type of plot that displays the relationship between two variables. In Seaborn, we can create a heat map using the heatmap() function.

```python
sns.heatmap(data=np.random.randn(10, 10))
plt.show()
```
